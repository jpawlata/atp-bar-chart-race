```python
import pandas as pd
import bar_chart_race as bcr
```

```python
# Get the data
df = pd.read_csv('https://media.githubusercontent.com/media/jpawlata/ATP_Ranking/main/data/atp_ranking.csv')
```

```python
# Take the first 10 rows for each date (15 top ATP rankings) from the dataset
df = df.groupby('Date').head(10).reset_index(drop = True)
```

```python
df.head(15)
```

```python
# Keep the important columns only
columns = ['Date', 'Player', 'Points']
df = df[columns]
```

```python
# Convert Date into datetime format
df.Date = pd.to_datetime(df.Date, format = "%Y-%m-%d")
```

```python
df.info()
```

```python
# Check null values
df.isnull().sum()
```

```python
# Drop rows with null Points
df = df.dropna(subset = ['Points'])
```

```python
# Check if any date has 0 points (that might be wrong data)
df.Points.isin([0, 0.0]).any()
```

```python
# Check the rows with 0.0 points
df.loc[df.Points.isin([0, 0.0])]
```

```python
# Remove rows with Points = 0.0
df = df[df.Points != 0.0]
```

```python
# Check
df.Points.isin([0, 0.0]).any()
```

```python
# Transform data for bar chart race
dfp = df.pivot(index = 'Date', columns = 'Player', values = 'Points')
```

```python
dfp.head()
```

```python
# Fill NaN with 0
dfp = dfp.fillna(0)
```

```python
dfp.head()
```

```python
# Create bar chart race
bcr.bar_chart_race(
    df = dfp, 
    filename = 'atp_ranking.mp4',
    fig_kwargs = {
        'figsize': (26, 15),
        'dpi': 120,
        'facecolor': '#FFFFFF'
    },
    title = {
        'label': 'TOP 10 - ATP Singles Ranking (1996 - 2021)',
        'size': 50, 
        'weight': 'bold',
        'pad': 30,
        'color': '#797979',
        'family': 'DejaVu Sans'
    },
    n_bars = 10, 
    fixed_max = False,
    bar_kwargs = {
        'alpha': .9, 
        'lw': 0.9
    },
    colors = 'pastel', 
    filter_column_colors = True,
    steps_per_period = 20,
    period_length = 600,
    tick_label_font = {
        'size': 20,
        'family': 'DejaVu Sans',
        'weight': 'bold',
        'color': '#797979'
    },
    bar_label_font = {
        'size': 20,
        'family': 'DejaVu Sans',
        'color': '#797979'
    },
    shared_fontdict = {
        'family': 'DejaVu Sans', 
        'color': '#797979',
        'size': 25
    },
    period_label = {
        'x': .95, 
        'y': .15, 
        'ha': 'right', 
        'va': 'center', 
        'size': 30,
        'color': '#797979',
        'weight': 'bold',
        'family': 'DejaVu Sans' 
    },
    period_template = f'Date:  ' + '%B %d, %Y' 
)
```
